```python
import numpy as np
import matplotlib.pyplot as plt
# How many time points are needed i,e., Sampling Frequency
fs= 500
# Create two sine waves
# Get x values of the sine wave
t=fs*0.01
time= np.arange(0, t,1/fs)
# Amplitude of the sine wave is sine of a variable like time
W1=4    #Frequency1
W2=10     #Frequency2
W3=7     #Frequency3
S1=0     # shifted wave1 (rigth or left)
S2=0     # shifted wave2 (rigth or left)
S3=0     # shifted wave3 (rigth or left)
A1=1     #amplitudet wave1
A2=2    #amplitudet wave1
A3=5    #amplitudet wave3
D1=0     # shifted wave1 up or down
D2=0     # shifted wave2 up or down
D3=0     # shifted wave3 up or down
sin1 = A1*np.sin(2*np.pi*W1*time+S1)+D1
sin2 = A2*np.sin(2*np.pi*W2*time+S2)+D2
sin3 = A3*np.sin(2*np.pi*W3*time+S3)+D3
# Time domain representation for sine wave 1
plt.figure(figsize = (20,5))
plt.plot(time, sin1)
plt.title('Sine wave with a frequency of 4 Hz and Amplitude = 1')
plt.xlabel('Time')
plt.ylabel('Amplitude = sin(time)')
plt.show()
# Time domain representation for sine wave 2
plt.figure(figsize = (20,5))
plt.plot(time, sin2)
plt.title('Sine wave with a frequency of 10 Hz and Amplitude = 2')
plt.xlabel('Time')
plt.ylabel('Amplitude = sin(time)')
plt.show()
# Time domain representation for sine wave 2
plt.figure(figsize = (20,5))
plt.plot(time, sin3)
plt.title('Sine wave with a frequency of 7 Hz and Amplitude = 5')
plt.xlabel('Time')
plt.ylabel('Amplitude = sin(time)')
plt.show()
# Add the sine waves
signal = sin1 + sin2 + sin3
# Time domain representation of the resultant sine wave
plt.figure(figsize = (20, 5))
plt.plot(time, signal)
plt.title('Sine wave with multiple frequencies')
plt.xlabel('Time')
plt.ylabel('Amplitude = sin(time)')
plt.show()

# Frequency domain representation
N=len(signal)
complex_sine_wave_t = np.arange(0,1,1/N)  # normalised time
fr = np.arange(0,N-1,1)  # to always have frequencies that are independent of the sampling rate
nyquist = fs/2
pts_hz = np.linspace(0,nyquist,int(((N/2)/t))) # cannot go higher to prevent aliasing
coeff=[]
for i in range(1,len(fr)):
    complex_sine_wave = np.exp(-2j*np.pi*fr[i]*complex_sine_wave_t); # stop it to see the magnitude
    coeff.append((np.dot(complex_sine_wave,signal))/N)
#coeff=dft(sin3)
m = np.abs(coeff)
m=m*2 #doubling the positive frequencies 

# Frequency domain representation
plt.figure(figsize = (20,5))
#plt.plot(frequencies, abs(fourierTransform))
plt.plot(pts_hz/t, m[0:len(pts_hz/t)])
plt.title('Fourier transform depicting the frequency components')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.show()
```
